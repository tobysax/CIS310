```python
#ticks, labels, and legends
fig = plt.figure()
ax = fig.add_subplot(111)
ax.plot(randn(1000).cumsum())
```

```python
ax.set_xticks([0,250,500,750,1000])

#ax.set_xtickslabels(['one','two','three','four','five'], rotation=30, fontsize='small')

ax.set_title('My First Matplotlip Plot with Ticks')

ax.set_xlabel('Stages')
ax.set_ylabel('Values')

fig
```

```python
#Use dictionary for title, xlabel, and ylabel
props = {'title': 'My First Matplotlip Plot with Ticks',
        'xlabel': 'Stages',
        'ylabel': 'Values'}

ax.set(**props)
fig
```

```python
fig = plt.figure()
ax = fig.add_subplot(1,1,1)

ax.plot(randn(1000).cumsum(), 'k-', label='One')
ax.plot(randn(1000).cumsum(), 'k--', label='Two')
ax.plot(randn(1000).cumsum(), 'k.', label='Three')

ax.legend(loc='best')
```
